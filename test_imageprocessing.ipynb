```python
import sagemaker
from sagemaker import get_execution_role
 
role = get_execution_role()
print(role)
sess = sagemaker.Session()
```

```python
#https://raz-sagemaker.s3.eu-central-1.amazonaws.com/data/xray/chest_xray/sample/test/IM-0033-0001-0002.jpeg

bucket = 'raz-sagemaker'
prefix = 'data/xray/chest_xray/sample/test/IM-0001-0001.jpeg'

import boto3
import tempfile
s3 = boto3.resource('s3', sess.boto_region_name)
bucket = s3.Bucket(bucket)
object = bucket.Object(prefix)
tmp = tempfile.NamedTemporaryFile()
print(object)
print(tmp.name)

with open(tmp.name, 'wb') as data:
    object.download_fileobj(data)

from PIL import Image, ImageFilter
#Read image
im = Image.open(tmp.name)
#Display image
im.show()

import matplotlib.pyplot as plt
plt.imshow(im)
plt.show()
```
